```python
import os
import numpy as np
import cv2
import matplotlib.pyplot as plt
import pickle
import random
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn import metrics

"""""
dir = 'C:\\Users\\car'
categories = ['front', 'Back']



data = []

for category in categories:
    path = os.path.join(dir, category)
    label = categories.index(category)

    for img in os.listdir(path):
        imgpath = os.path.join(path, img)
        car_img = cv2.imread(imgpath, 0)
        try:
            car_img = cv2.resize(car_img , (50,50))
            image = np.array(car_img).flatten()

            data.append([image, label])
        except Exception as e:
            pass

pick_in = open('dataCar.pickle', 'wb')
pickle.dump(data, pick_in)
pick_in.close()

"""""

pick_in = open('dataCar.pickle', 'rb')
data = pickle.load(pick_in)
pick_in.close()


random.shuffle(data)
features = []
labels = []

for feature, label in data:
    features.append(feature)
    labels.append(label)


xtrain, xtest, ytrain, ytest = train_test_split(features, labels, test_size=0.1)

#model = SVC(C=1.0, kernel='poly', gamma='auto')
#model.fit(xtrain, ytrain)

pick = open('modelCar.sav', 'rb')
model = pickle.load(pick)
#pickle.dump(model, pick)
pick.close()

prediction = model.predict(xtest)
accuracy = model.score(xtest, ytest)
categories = ['front', 'Back']

print('Accuracy: ', accuracy)
print('Prediction is: ', categories[prediction[0]])

mycar = xtest[0].reshape(50, 50)
plt.imshow(mycar, cmap='gray')
plt.show()
```
